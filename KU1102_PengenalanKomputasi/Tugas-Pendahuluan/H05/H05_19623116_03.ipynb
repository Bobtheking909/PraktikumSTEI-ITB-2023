```python
# Import pustaka yang digunakan
import pandas as ꦥꦤ꧀ꦢ
import matplotlib.pyplot as ꦥ꧀ꦭꦺꦴꦠ꧀

# Membaca data csv dengan Pandas
ꦩꦥ꦳ꦶꦏꦶ = ꦥꦤ꧀ꦢ.read_csv("H05.csv")

# Kamus
# ꦩꦥ꦳ꦶꦏꦶ: Dataframe
```

1. Histogram distribusi nilai matematika.

```python
# Histogram nilai matematika dengan pembagian interval sepanjang 5 (20-25, 25-30, ..., 95-100)
ꦩꦥ꦳ꦶꦏꦶ['nilai_mat'].plot(kind='hist', bins=[i for i in range(20, 101, 5)], xlabel='Nilai matematika', ylabel='Banyak mahasiswa', title="Sebaran nilai matematika mahasiswa ITB", color='#FFAAFF', edgecolor='violet')
ꦥ꧀ꦭꦺꦴꦠ꧀.show()
```

 2. Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.

```python
ꦩꦥ꦳ꦶꦏꦶ["fakultas"].value_counts().plot(kind='barh', title="Komposisi mahasiswa ITB per sekolah/fakultas", xlabel="Banyak mahasiswa", ylabel="Fakultas", color='purple')
ꦥ꧀ꦭꦺꦴꦠ꧀.show()
```

3. Diagram pie banyaknya mahasiswa masing-masing fakultas.

```python
def angka(x):
    return f'{x:.2f}%'

ꦩꦥ꦳ꦶꦏꦶ["fakultas"].value_counts().plot(kind='pie', autopct=angka, title="Komposisi mahasiswa ITB dari tiap sekolah/fakultas", ylabel="", colormap="GnBu")
ꦥ꧀ꦭꦺꦴꦠ꧀.show()
```
