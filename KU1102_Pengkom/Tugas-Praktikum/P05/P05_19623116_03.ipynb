```python
#include <pandas.h>
import pandas as plt
#include <matplotlib.pyplot.h>
import matplotlib.pyplot as pd

# Import data .csv
df = plt.read_csv("P05-03.csv")

# Kamus
# df    : Dataframe
```

1. Histogram distribusi banyaknya review di aplikasi yang memiliki jenis penyediaan ruang Private room dengan bins 30.

```python
df["number_of_reviews"][df["room_type"] == "Private room"].plot(kind="hist", bins=[i for i in range(0, 201, 20)])
pd.show()
```

2. Pie chart perbandingan banyaknya masing-masing jenis penyediaan ruang dengan banyaknya review di aplikasi di atas 100.

```python
df["room_type"][df["number_of_reviews"] > 100].value_counts().plot(kind="pie")
pd.show
```

3.  Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang tidak dimiliki oleh Jay.

```python
# Saring data sehingga hanya berisi rumah yang tidak dimiliki oleh Jay: df[df["host_name"] != "Jay"]
df[df["host_name"] != "Jay"].plot(kind="scatter", x="longitude", y="latitude", color="orange")
pd.show()
```

4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya hari ketersediaan rumah dalam satu tahun sebagai ukuran dari rumah yang berada di neighbourhood Novena.

```python
# Saring data sehingga hanya berisi rumah di neighbourhood Novena: df[df["neighbourhood"] == "Novena"]
df[df["neighbourhood"] == "Novena"].plot(kind="scatter", x="longitude", y="latitude", sizes=df["availability_365"], color="violet")
pd.show()
```

5. Stacked bar chart dari rumah yang disewakan oleh ’JoJo’, ’Elizabeth’, dan ’Jimmy’, dengan room type sebagai x dan nama pemilik rumah sebagai stacked y.

```python
# Saring data sehingga hanya berisi rumah yang disewakan JoJo, Elizabeth, dan Jimmy
df = df[(df["host_name"] == "JoJo") | (df["host_name"] == "Elizabeth") | (df["host_name"] == "Jimmy")]
# Kelompokkan data berdasarkan `host_name`dan `room_type`
df = df.groupby(["room_type", "host_name"])["id"].size().unstack()

# Plot stacked bar chart
df.plot(kind="bar", y= ["Elizabeth", "Jimmy", "JoJo"], stacked=True)
pd.show()

# Entah kenapa plotnya hanya muncul kalau kita run all, sedangkan run cell ini malah error di laptop saya kak :(
```
